```python
import pandas as pd
```

```python
df_2016 = pd.read_excel('2016_poblanos_state.xlsx')
```

```python
df_2016 = df_2016.sort_values(by= 'state')
```

```python
df_2015 = pd.read_excel('2015_poblanos_state.xlsx')
df_2015 = df_2015.sort_values(by= 'state')
```

```python
df_2014 = pd.read_excel('2014_poblanos_state.xlsx')
df_2014 = df_2014.sort_values(by= 'state')
```

```python
df_2013 = pd.read_excel('2013_poblanos_state.xlsx')
df_2013 = df_2013.sort_values(by= 'state')
```

```python
df = pd.merge(df_2016, df_2015, on='state')
df
```

```python
df1 = pd.merge(df, df_2014, on='state')
```

```python
df2 = pd.merge(df1, df_2013, on='state')
df2
```

```python
df_total = df2[['state', 'amount_2016', 'amount_2015', 'amount_2014', 'amount_2013']]
df_total
```

```python
df_total = df_total.append(df_total.sum(numeric_only=True), ignore_index=True)
```

```python
df_total
```

```python
df_total['Total'] = df_total.amount_2016 + df_total.amount_2015 + df_total.amount_2014 + df_total.amount_2013
```

```python
df_total
```

```python
df_total.sort_values(by= 'Total', ascending = False)
```
